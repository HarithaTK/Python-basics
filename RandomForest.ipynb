```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python

# Importing Data 
from sklearn.datasets import load_boston 
boston = load_boston() 
 

```

```python
data = pd.DataFrame(boston.data) 
data.columns = boston.feature_names 
  
data.head(10)
```

```python
# Adding 'Price' (target) column to the data  

data['Price'] = boston.target 
data.head() 
```

# Scaling down the features:

```python
feature_scale= [feature for feature in data.columns if feature not in ['Price']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(data[feature_scale])
```

```python
scaler.transform(data[feature_scale])
```

```python
#adding back the columns for train dataset:

data=pd.concat([data[['Price']].reset_index(drop=True),
                pd.DataFrame(scaler.transform(data[feature_scale]),columns=feature_scale)],axis=1)
```

```python
data.head()
```

```python
y=data[['Price']]
```

```python
X=data[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
```

```python
X.head()
```

```python
y.head()
```

# Random Forest:Build Model


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfregressor=RandomForestRegressor()
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
rfregressor.fit(X_train,y_train)
```

```python
print('Coeeficient of determination R^2 for train data :{}'.format(rfregressor.score(X_train,y_train)))
```

```python
print('Coeeficient of determination R^2 for test data :{}'.format(rfregressor.score(X_test,y_test)))
```

```python
#Interpretation:From the above R^2 values we can clearly see that it is agood model

#We have very less variance between the R^2 value of train and test data.
```
